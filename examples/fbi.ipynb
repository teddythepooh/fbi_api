## Example

The Originating Agency Identifer (ORI) uniquely identifies the 19,000+ law enforcement agencies in the FBI data. If no api_key is passed in `FBI()`, your environment variable `FBI_API_KEY` is automatically invoked. 

- `get_metadata()`: Extract metadata of agencies by state, including but not limited to the ORIs.
- `get_agency_metrics()`: Extract the number of sworn and non-sworn officers in the police agency (and the total population they serve). The `ori` and `year` parameters accept both a string and a list of strings. 
- `get_crime_statistics()`: Extract monthly crime statistics. The `ori`, `year`, and `offense` parameters accept both a string and a list of strings. 

```python
from fbi_api import FBI

api = FBI()
```

```python
ori = "NY0303000"
offenses = ["Violent Crimes", "Aggravated Assault", "Burglary", "Larceny"]

metadata = api.get_metadata("all")

agency_metrics = api.get_agency_metrics(ori = ori, year = 2024)

nyc_crime_statistics = api.get_crime_statistics(ori = ori, year = [2022, 2024], offense = offenses)
```
